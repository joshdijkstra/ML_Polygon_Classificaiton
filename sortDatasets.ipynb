```python
import pandas as pd
import os
import shutil
```

```python
pwd = os.getcwd()
targets_file = str(pwd) + '\\targets.csv'

data = pd.read_csv(targets_file)
data
```

```python
shapes = {3 : 'triangle' , 4 : 'square' , 5 : 'pentagon' , 6 : 'hexagon' }
data['sides'] = data['sides'].map(shapes)
data
```

```python
filenames = data['filename'].values.tolist()
sides = data['sides'].values.tolist()
```

```python
for x in range(len(filenames)):
    fileloc = str(pwd) + "\\images\\content\\" + filenames[x]
    if (sides[x] == 'triangle'):
        target = str(pwd) + "\\images\\content\\triangle\\"
    elif (sides[x] == 'square'):
        target = str(pwd) + "\\images\\content\\square\\"
    elif (sides[x] == 'pentagon'):
        target = str(pwd) + "\\images\\content\\pentagon\\"
    else:
        target = str(pwd) + "\\images\\content\\hexagon\\"    
    shutil.move(fileloc, target + filenames[x])


```
